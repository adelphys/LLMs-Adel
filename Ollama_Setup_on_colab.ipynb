# Using Python Library:
We show how to interact with the Ollama server using its Python client library.
The ollama library, initializes a client to connect to the running Ollama server, and then uses that client to generate a response from the llama3.2:3b model based on a given prompt. Finally, it prints the model's generated response.

```python
# Install the library first
!pip install ollama -q
import ollama

# Instantiate the Ollama client (by default it connects to 127.0.0.1:11434)
client = ollama.Client()
# Generate a response from the model
response = client.generate(model='llama3.2:3b', prompt='Summarize the benefits of local AI in one sentence.')
print(response['response'])

# or we can use
# response = ollama.chat( model='llama3.2:3b', m essages=[ {'role': 'user', 'content': 'Why is Google Colab useful for AI?'}]  )
```

# with CUDA support

---


Below is a complete, one-cell script that installs Ollama, starts the server with CUDA support, pulls a science-capable model (like Llama 3.2).

**Prerequisites**
Before running the script, ensure your Colab runtime is set to GPU:

*   Go to Runtime > Change runtime type.
*   Select T4 GPU (or any available GPU) and click Save.


---



*To run Ollama on the CPU, you can set the OLLAMA_NO_GPU=1 environment variable, in run_ollama_serve() function, before starting the Ollama server. This will prevent Ollama from using any available GPU and force it to run on the CPU. *



```python
# 1. Install Ollama and dependencies
!sudo apt-get update
!sudo apt-get install zstd
!sudo dnf install zstd
!sudo pacman -S zstd
!curl -fsSL https://ollama.com/install.sh | sh

# 2. Start Ollama server in the background with CUDA support
import subprocess
import threading
import time
import os

def run_ollama_serve():
    # Setting environment variable to ensure CUDA is prioritized
    os.environ['OLLAMA_INTEL_GPU'] = "0"
    subprocess.Popen(["ollama", "serve"])

# Use threading to keep the server running while we use the notebook
thread = threading.Thread(target=run_ollama_serve)
thread.start()
time.sleep(10) # Give the server time to initialize

# 3. Install the Ollama Python library and pull the model
!pip install ollama
import ollama

# Pulling Llama 3.2 (3B) - good balance of speed and reasoning for science
MODEL = "llama3.2"
print(f"Pulling {MODEL}... this may take a minute.")
ollama.pull(MODEL)

# 4. Define the solver function
def Summarize_text_LLM_model(insert_text):
    system_prompt = ( "Summarize the benefits of local AI in one sentence." )

    response = ollama.generate( model=MODEL, system=system_prompt, prompt=insert_text )
    return response['response']


```

# test the example usage:

```python
# --- EXAMPLE USAGE ---

text = "Summarize the benefits of local AI in one sentence."
print("\n--- sumurized text ---")
print(Summarize_text_LLM_model(text))
```

## Running Ollama on CPU (No GPU)

```python
# 1. (Optional) Stop any running Ollama server to avoid conflicts
!kill $(lsof -t -i:11434)

import subprocess
import threading
import time
import os

def run_ollama_serve_cpu():
    # Set environment variable to explicitly disable GPU
    os.environ['OLLAMA_NO_GPU'] = '1'
    subprocess.Popen(["ollama", "serve"])

# Use threading to keep the server running in the background
thread_cpu = threading.Thread(target=run_ollama_serve_cpu)
thread_cpu.start()
time.sleep(10) # Give the server time to initialize
print("ðŸš€ Ollama server started in CPU-only mode!")
```

Once the CPU-only server is running, you can pull and use models just like before. Smaller models are recommended for better performance on CPU.

```python
# Install the Ollama Python library if not already installed
!pip install ollama -q

import ollama

# Pull a small model for CPU (e.g., Phi-3 Mini)
CPU_MODEL = "phi3:mini"
print(f"Pulling {CPU_MODEL} for CPU... this may take a minute.")
ollama.pull(CPU_MODEL)
```

```python
# Test the CPU-only model
response_cpu = ollama.generate(model=CPU_MODEL, prompt='Explain the concept of 'AI hallucination' in one sentence.')
print("\n--- CPU-only Model Response ---")
print(response_cpu['response'])
```

## Extracting Text from a PDF File

To process PDF files, we'll use the `PyMuPDF` library (often imported as `fitz`).

```python
# Install PyMuPDF
!pip install PyMuPDF -q
```

### Upload your PDF file

Before proceeding, please upload your PDF file to the Colab environment. You can do this by clicking the 'Files' icon on the left sidebar, then clicking the 'Upload to session storage' icon, and selecting your PDF file. Make sure to note down the filename.

```python
import fitz  # PyMuPDF
import os

# --- USER INPUT REQUIRED ---
# Replace 'your_document.pdf' with the actual name of your uploaded PDF file.
pdf_filename = 'your_document.pdf' # @param {type:"string"}
# --- END USER INPUT ---


if not os.path.exists(pdf_filename):
    print(f"Error: File '{pdf_filename}' not found. Please upload the PDF file and ensure the filename is correct.")
else:
    try:
        # Open the PDF file
        document = fitz.open(pdf_filename)
        full_text = ""

        # Iterate through each page and extract text
        for page_num in range(len(document)):
            page = document.load_page(page_num)
            full_text += page.get_text()
            full_text += "\n\n" # Add some separation between pages

        # Print the extracted text (first 1000 characters for brevity)
        print("--- Extracted Text (first 1000 characters) ---")
        print(full_text[:1000])

        # You can also save the full text to a variable for further processing
        # extracted_pdf_text = full_text

    except Exception as e:
        print(f"An error occurred while processing the PDF: {e}")
```
